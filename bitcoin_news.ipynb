```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# Initialize the DataFrame before the loop
df = pd.DataFrame(columns=["date", "title"])

page_num = 1
proceed = True

while page_num < 3:
    # Construct the URL with the correct page number
    url = f"https://www.ft.com/search?q=bitcoin&page={page_num}&dateTo=2024-11-07&dateFrom=2024-05-01&sort=relevance&isFirstView=false"
    
    # Fetch the page
    response = requests.get(url)
    soup = BeautifulSoup(response.text, "html.parser")  # Use "html.parser" correctly
    
    # # Check if page is 404
    # if soup.title and soup.title.text == "404 Not Found":
    #     proceed = False
    # else:
        # Find titles and dates
    titles = soup.find_all('div', class_='o-teaser__heading')
    dates = soup.find_all('div', class_='o-teaser__timestamp')
    
    # Extract and store data
    for title, date in zip(titles, dates):
        text = title.find('a').get_text(strip=True)
        date_text = date.find('time').get_text(strip=True)
        length = len(df)
        
        # Assign data to the DataFrame
        df.loc[length, "date"] = date_text
        df.loc[length, "title"] = text
    
    # Increment page number
    page_num += 1
    print(page_num)

# Display the resulting DataFrame
df
```
