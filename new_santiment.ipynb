```python
import numpy as np
import pandas as pd
import scipy
import nltk
from transformers import pipeline
```

```python
# import torch  # If using PyTorch

# Initialize the sentiment analysis pipeline
sentiment_analyzer = pipeline("sentiment-analysis")


transformer_nlp = pd.read_csv("bitcoin_news.csv")

# Apply the sentiment analysis model to each title
transformer_nlp['sentiment'] = transformer_nlp['title'].apply(lambda x: sentiment_analyzer(x)[0]['label'])

transformer_nlp
```

```python
from nltk.sentiment import SentimentIntensityAnalyzer
import pandas as pd
import nltk

# Download VADER if you haven't yet
nltk.download('vader_lexicon')

# Initialize VADER sentiment analyzer
sia = SentimentIntensityAnalyzer()


df = pd.read_csv("bitcoin_news.csv")

# Apply VADER sentiment analysis
def classify_sentiment_vader(text):
    score = sia.polarity_scores(text)['compound']
    return 'positive' if score > 0 else 'negative'

df['sentiment'] = df['title'].apply(classify_sentiment_vader)

print(df)
```
